```python
#ATM SIMULATION - 04 APRIL 2020
# DEFINING AND INITIALIZING ALL VARIABLES
restart = ('y')
chances = 3
withdrawAmount = 0
pay_in = 0
balance = 2020.04
withdrawBal = (balance - withdrawAmount)
depositBal = (balance + pay_in)
mainBal = balance
# START OF CODE
print('Wecome to Corona Bank ATM')
while chances >= 0:
    pin = int(input('Please Enter Your 4 Digit Pin:\n'))     # SETTING PIN TO 1234
    if pin == (1234):
        print('Pin Correct! Please proceed...\n')
        while restart == ('y'):                              # TRANSACTION OPTIONS
            print('Press 1 for Account Balance\n')
            print('Press 2 for Withdrawal\n')
            print('Press 3 for Deposit\n')
            print('Press 4 to End Transaction\n')
            option = int(input('Please make your choice\n')) # MAKE A CHOICE FROM THE OPTIONS
            if option == 1:                                  # ITERATIONS TO DISPLAY CURRENT ACCOUNT BALANCE
                if pay_in is True:
                    mainBal = depositBal
                    if withdrawAmount is True:
                        mainBal = withdrawBal
                print('Your account balance is N:', mainBal,'\n')  # DISPLAY BALANCE ANY TIME OPTION IS SELECTED
                if mainBal <= 0:                                   # MAKE SURE WE DO NOT HAVE A NEGATIVE BALANCE
                    print('Insufficient funds. Please enter a smaller amount')
                    break
                restart = input('Would you like to make another transaction? (Y/N)') # MAKE A NEW TRANSACTION
                if restart in ('n','N'):                                             # END TRANSACTION
                    print('Thank you for banking with us')
                    print('Please take your card')
                    break
            elif option == 2:                                                                       # ITERATIONS FOR WITHDRAWAL
                withdrawAmount = float(input('Enter Amount to withdraw \n N1000/N2000/N5000/N10000/N20000/40000\n'))                   
                if withdrawAmount in [1000, 2000, 5000, 10000, 20000, 40000]:                       # WITHDRAWAL OPTIONS  
                    withdrawBal = (mainBal - withdrawAmount)         
                    if withdrawAmount >= mainBal:                                            # PREVENTING OVERWITHDRAWAL
                        print('Insufficient funds. Please enter a smaller amount')
                        withdrawBal = (withdrawBal + withdrawAmount)
                        break                        
                    if withdrawAmount < mainBal:
                        print('Your account balance is N',withdrawBal,'\n')
                        mainBal = withdrawBal
                    restart = input('Would you like to make another transaction? (Y/N)')
                    if restart in ('n','N'):
                        print('Thank you for banking with us')
                        print('Please take your card')
                        break        
                else:                            
                    print('invalid amount entered, Please try again\n')
                    restart = ('y')             
            elif option == 3:                                                # ITERATIONS FOR DEPOSIT
                if pay_in is True:
                    mainBal = balance
                pay_in = float(input('How much would you like to deposit?')) # NO RESTRICTIONS ON DEPOSIT AMOUNT
                depositBal = mainBal + pay_in
                mainBal = depositBal
                print('Your account balance is now N',mainBal,'\n')
                restart = input('Would you like to make another transaction? (Y/N)')
                if restart in ('n','N'):
                    print('Thank you for banking with us')
                    print('Please take your card')
                    break
            elif option == 4:                                               # TO END THE TRANSACTION
                print('Thank you for banking with us')
                print('Please take your card')
                break
        else:
            print('Please enter the correct Pin')
            restart = ('y')  
        if pin != (1234):                                                  # PIN CHECK!
            print('Incorrect Pin')
            chances = chances - 1                                          # CHECK NUMBER OF INCORRECT PIN ATTEMPT
            if chances == 0:
                print('You have entered an incorrect Pin 3 times, transaction is now terminated')
                break
```
